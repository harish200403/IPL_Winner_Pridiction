### Now lets merge this dataset into the dataset match

```python
match = match.merge(total_score[['match_id','total_runs']],left_on='id',right_on='match_id')
# in the match dataset, merge match_id and total_runs columns of total_score
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals']

#These teams play presently
```

```python
match['team1'] = match['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match['team2'] = match['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
match['team1'] = match['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match['team2'] = match['team2'].str.replace('Deccan Chargers','Sunrisers Hyderaba')
```

```python
match = match[match['team1'].isin(teams)]
match = match[match['team2'].isin(teams)]
# team1 me unhi teams ka match rakho jo 'teams' me hai
# team2 me unhi teams ka match rakho jo 'teams' me hai
```

```python
match['dl_applied'].value_counts()
#dl =  Duckworth Lewis
```

```python
match = match[match['dl_applied']==0]
# means jisme jisme dl applied hua hai wo hata do
```

```python
match = match[['match_id','city','winner','total_runs']]
```

### Now lets merge delivery and match

```python
delivery=match.merge(delivery,on='match_id')
```

```python
delivery
```

```python
delivery=delivery[delivery['inning']==2]
```

```python
# Assuming 'total_runs_y' column contains numeric values stored as strings, convert them to integers
delivery['total_runs_y'] = delivery['total_runs_y'].astype(int)

# Now, you can use the cumsum() function on the converted 'total_runs_y' column
delivery['current_score'] = delivery.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery['runs_left'] = delivery['total_runs_x'] - delivery['current_score']
```

```python
delivery['balls_left'] = 126 - (delivery['over']*6 + delivery['ball'])
```

```python
delivery
```

```python
# Fill missing values in 'player_dismissed' column with "0"
delivery['player_dismissed'] = delivery['player_dismissed'].fillna("0")

# Convert 'player_dismissed' column to boolean: True if dismissed, False otherwise
delivery['player_dismissed'] = delivery['player_dismissed'] != "0"

# Calculate cumulative wickets for each match
wickets = delivery.groupby('match_id')['player_dismissed'].cumsum()

# Calculate remaining wickets
delivery['wickets'] = 10 - wickets
```

```python
delivery.head()
```

```python
# CRR = Curretnt Run Rate = runs/overs
delivery['CRR'] = (delivery['current_score']*6)/(120 - delivery['balls_left'])
```

```python
# RRR = Required run rate
delivery['RRR'] = (delivery['runs_left']*6)/delivery['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery['result'] = delivery.apply(result,axis=1)
```

```python
final = delivery[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','CRR','RRR','result']]
```

```python
final.head()
```

```python
final = final.sample(final.shape[0])
# It will do suffle
```

```python
final.sample()
```

```python
final.dropna(inplace=True)
```

```python
final = final[final['balls_left'] != 0]
```

```python
X = final.iloc[:,:-1]
y = final.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

### Applying one hot encoding on batting_team,bowling_team,city

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
model = LogisticRegression(solver='liblinear')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',model)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(X_test)[1]
```
